## 3. Cluster Analysis

The main purpose of this project is to perform [cluster analysis](https://en.wikipedia.org/wiki/Cluster_analysis#:~:text=Cluster%20analysis%20or%20clustering%20is,in%20other%20groups%20(clusters).) with the [K-Means](https://towardsdatascience.com/understanding-k-means-clustering-in-machine-learning-6a6e67336aa1) algorithm. 

You can perform many [cluster analysis](http://www.stat.columbia.edu/~madigan/W2025/notes/clustering.pdf) using different variables. If you use a maximum of two variables for each cluster analysis, you can identify cluster labels more clearly.

First, the K-Means algorithm expects you to determine the number of clusters (*n_clusters*). You can determine the optimal number of clusters for each cluster analysis in various ways. In this case, you are expected to use the [Elbow Method](https://en.wikipedia.org/wiki/Elbow_method_(clustering).

Finally, different information is obtained in each analysis. Therefore, different labeling should be done as a result of each cluster analysis. 

Labeling example: 

- **Normal Customers**  -- An Average consumer in terms of spending and Annual Income
- **Spender Customers** --  Annual Income is less but spending high, so can also be treated as potential target customer.

### Clustering based on Age and Spending Score

#### *i. Create a new dataset with two variables of your choice*

```python
df_1 = df[["Age", "Spending Score (1-100)"]]
X_1 = df_1
X_1
```

```python
from pyclustertend import hopkins
from sklearn.preprocessing import scale
```

```python
X_1.shape
```

```python
hopkins(scale(X_1), X_1.shape[0])
```

```python
hopkins(X_1, X_1.shape[0])
```

#### *ii. Determine optimal number of clusters*

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k)
    model.fit(X_1)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(X_1)
visu.poof()
```

#### *iii. Apply K Means*

```python
model = KMeans(n_clusters =4)
model.fit(X_1)
model
```

#### *iv. Visualizing and Labeling All the Clusters*

```python
labels = model.labels_
labels
```

```python

X_1["cluster_id"] = labels
```

```python
X_1.head()
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
print(f'Silhouette Score(n=4): {silhouette_score(X_1, labels)}')
```

```python
plt.scatter(X_1.iloc[:,0], X_1.iloc[:,1], c = labels, s = 50, cmap = "viridis")
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, c='black',alpha=0.5, label = 'Centroids')
```

### Clustering based on Annual Income and Spending Score

#### *i. Create a new dataset with two variables of your choice*

```python
df_2 = df[["Annual Income (k$)", "Spending Score (1-100)"]]
X_2 = df_2
X_2
```

```python
hopkins(X_2, X_2.shape[0])
```

```python
hopkins(scale(X_2), X_2.shape[0])
```

#### *ii. Determine optimal number of clusters*

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k)
    model.fit(X_2)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
```

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(scale(X_2))
visu.poof()
```

#### *iii. Apply K Means*

```python
model = KMeans(n_clusters =5)
model.fit(X_2)
model
```

#### *iv. Visualizing and Labeling All the Clusters*

```python
labels = model.labels_
labels
```

```python
X_2["cluster_id"] = labels
```

```python
X_2.head()
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
print(f'Silhouette Score(n=5): {silhouette_score(X_2, labels)}')
```

```python
plt.scatter(X_2.iloc[:,0], X_2.iloc[:,1], c = labels, s = 50, cmap = "viridis")
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, c='black',alpha=0.5, label = 'Centroids')
```

### Hierarchical Clustering

#### *i. Determine optimal number of clusters using Dendogram*

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
hc_ward = linkage(X_1, "ward")
```

```python
plt.figure(figsize =(17,8))
plt.title("dendogram")
plt.xlabel("Observation")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10);
```

```python
hc_ward = linkage(X_2, "ward")
```

```python
plt.figure(figsize =(15,8))
plt.title("dendogram")
plt.xlabel("Observation")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10);
```

#### ii. *Apply Agglomerative Clustering*

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters =4, affinity = "euclidean", linkage = "ward")
model.fit_predict(X_1)
```

```python
model.labels_
```

#### iii. *Visualizing and Labeling All the Clusters* 

```python
plt.scatter(X_1.iloc[:,0], X_1.iloc[:,1], c = model.labels_, cmap = "rainbow")
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
print(f'Silhouette Score(n=4): {silhouette_score(X_1, model.labels_)}')
```

```python
model_2 = AgglomerativeClustering(n_clusters =5, affinity = "euclidean", linkage = "ward")
model_2.fit_predict(X_2)
```

```python
model_2.labels_
```

```python
plt.scatter(X_2.iloc[:,0], X_2.iloc[:,1], c = model_2.labels_, cmap = "rainbow")
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
print(f'Silhouette Score(n=5): {silhouette_score(X_2, model_2.labels_)}')
```

### Conclusion

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(X_1)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X_1, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(X_2)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X_2, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    # intialise kmeans
    AC = AgglomerativeClustering(n_clusters =num_clusters)
    AC.fit_predict(X_1)
    
    cluster_labels = AC.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X_1, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    # intialise kmeans
    AC = AgglomerativeClustering(n_clusters =num_clusters)
    AC.fit_predict(X_2)
    
    cluster_labels = AC.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X_2, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```
